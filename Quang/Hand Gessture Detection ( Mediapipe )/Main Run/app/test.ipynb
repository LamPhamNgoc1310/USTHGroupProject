```python
from tensorflow.keras.models import load_model
import mediapipe as mp
import cv2 as cv
import copy
from functions import *
```

```python
args = get_args()
cap_device = args.device
cap_width = args.width
cap_height = args.height   

use_static_image_mode = args.use_static_image_mode
min_detection_confidence = args.min_detection_confidence
min_tracking_confidence = args.min_tracking_confidence
```

```python

mp_hands = mp.solutions.hands
hands = mp_hands.Hands(
        static_image_mode=use_static_image_mode,
        max_num_hands=1,
        min_detection_confidence=min_detection_confidence,
        min_tracking_confidence=min_tracking_confidence,
    )
```

```python
# """Camera preparation"""
# cap = cv.VideoCapture(cap_device)
# cap.set(cv.CAP_PROP_FRAME_WIDTH, cap_width)
# cap.set(cv.CAP_PROP_FRAME_HEIGHT, cap_height)
# print(cap_device, cap_width, cap_height)
```

```python
model = load_model('keypoint_classifier_demo_2.keras')
```

```python
labels = [0,1,2,3,4,5]
cap = cv.VideoCapture(0)
while True:
    # print(1)
    """Process Key (ESC: end)"""
    key = cv.waitKey(10)
    if key == 27:  # ESC
        break
    
    # print(2)
    ret, image = cap.read()
    if not ret:
        break
    image = cv.flip(image, 1) # Mirror display ()
    debug_image = copy.deepcopy(image)

    # print(3)

    results = hands.process(image)


    # print(4)
    if results.multi_hand_landmarks is not None:
        for hand_landmarks, handedness in zip(results.multi_hand_landmarks, results.multi_handedness):
            
            # print(4.1)
            # Bounding Box calculation
            brect = calc_bounding_rect(debug_image, hand_landmarks)


            # print(4.2)
            # Landmark calculation
            Landmark_list = calc_landmark_list(debug_image, hand_landmarks)


            # print(4.3)
            # Conversion to relative coordinates / normalized coordinates
            pre_processed_landmark_list = pre_process_landmark(Landmark_list)

            # print(4.4)

            try:
                # Hand sign classification
                # hand_sign_id, confidence = keypoint_classifier.predict(pre_processed_landmark_list)
                pred = model.predict(np.array([pre_processed_landmark_list]), verbose=0)
                max_position = np.argmax(pred)
                pred_output = labels[max_position]
                cLevel = np.ceil(pred[0][max_position] * 100) / 100
                print(pred_output, cLevel)

            except: 
                print("fucked")
                print(pre_processed_landmark_list)
                print(len(pre_processed_landmark_list))
            # print(4.5)

    cv.imshow('Hand Gesture Recognition', image)
cap.release()
cv.destroyAllWindows()
```
