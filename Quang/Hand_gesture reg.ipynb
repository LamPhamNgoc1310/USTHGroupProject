# BEST VERSION 2 (Running Average model – Background Subtraction and segmented hand)

## CONS:
- CANNOT APPLY TRACKING METHOD BECAUSE OF THE RUNNING AVERAGE REQUIRE A FIXED FRAME, NOT THE MOTION FRAME

```python
import imutils
bg = None
```

```python
# Function - To find the running average over the background
def run_avg(image, accumWeight):
    global bg
    if bg is None:
        bg = image.copy().astype("float")
        return
    cv2.accumulateWeighted(image, bg, accumWeight)
```

```python
# Function - To segment the region of hand in the image
def segment(image, threshold=30):
    global bg
    diff = cv2.absdiff(bg.astype("uint8"), image)
    thresholded = cv2.threshold(diff, threshold, 255, cv2.THRESH_BINARY)[1]
    cnts, _ = cv2.findContours(thresholded.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(cnts) == 0:
        return
    else:
        segmented = max(cnts, key=cv2.contourArea)
        return (thresholded, segmented)
```

```python
# Function - To predict the class of the segmented hand region
def predict_gesture(thresholded):
    # Resize the image to match the model's expected input size
    resized = cv2.resize(thresholded, (64, 64))  # Adjust size based on your model's input shape
    
    # Convert the grayscale image to RGB by stacking the grayscale image three times along the last axis
    reshaped = np.stack((resized,) * 3, axis=-1)  # Now reshaped has shape (64, 64, 3)
    
    # Reshape to match the model input shape and normalize
    reshaped = reshaped.reshape(1, 64, 64, 3)  # Add batch dimension
    reshaped = reshaped.astype("float32") / 255.0  # Normalize pixel values to [0, 1]

    # Perform the prediction
    pred = model.predict(reshaped)
    id_class = np.argmax(pred)
    name_class = labels[id_class].capitalize()
    return name_class
```

```python
# Main function
if __name__ == "__main__":
    accumWeight = 0.5
    camera = cv2.VideoCapture(0)
    top, right, bottom, left = 10, 350, 400, 600
    num_frames = 0

    while True:
        (grabbed, frame) = camera.read()
        frame = imutils.resize(frame, width=700)
        frame = cv2.flip(frame, 1)
        clone = frame.copy()
        (height, width) = frame.shape[:2]
        roi = frame[top:bottom, right:left]

        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        gray = cv2.GaussianBlur(gray, (7, 7), 0)

        if num_frames < 30:
            run_avg(gray, accumWeight)
            if num_frames == 1:
                print(">>> Please wait! calibrating...")
            elif num_frames == 29:
                print(">>> Calibration successful...")
        else:
            hand = segment(gray)
            if hand is not None:
                (thresholded, segmented) = hand
                cv2.imshow("Thresholded", thresholded)

                # Predict the gesture
                gesture_name = predict_gesture(thresholded)
                
                # Display the prediction on the video feed
                cv2.putText(clone, f'Gesture: {gesture_name}', (50, 50), 
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Draw the segmented hand ROI
        cv2.rectangle(clone, (left, top), (right, bottom), (0, 255, 0), 2)
        num_frames += 1
        cv2.imshow("Video Feed", clone)

        keypress = cv2.waitKey(1) & 0xFF
        if keypress == ord("q"):
            break

# Free up memory
camera.release()
cv2.destroyAllWindows()
```
